```python
plt.figure(figsize=(20,10))
sns.set_style('whitegrid')
sns.boxplot(data=df)
```

```python
stats_df = df.drop(['Total', 'Stage', 'Legendary'], axis=1)
melted_df = pd.melt(stats_df,
                   id_vars=['Name', 'Type 1', 'Type 2'],
                   var_name = "Stat")
melted_df.head()
```

```python
plt.figure(figsize=(20,10))
sns.swarmplot(x='Stat', y='value', data=melted_df, hue='Type 1')
```
